## Seperate tokens to title and text column

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
df['title']= [word_tokenize(item) for item in df['title']]
df['text']= [word_tokenize(item) for item in df['text']]
df.head()
```

```python
df['text'][0]
```

## lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
# nltk.download('wordnet')
# nltk.download('stopwords')

def laama(data):
    return [lemmatizer.lemmatize(word) for word in data]
```

```python
df['text']=df['text'].apply(laama)
df['title']=df['title'].apply(laama)
```

## Removing stop words

```python
from nltk.corpus import stopwords
# nltk.download('stopwords')
stop = stopwords.words('english')

#Remove them from our dataframe
df['title'] = df['title'].apply(lambda x: [i for i in x if i not in stop])
df['text'] = df['text'].apply(lambda x: [i for i in x if i not in stop])


df.head()
```

## Converting tokens to string form

```python
df['title'] = df['title'].apply(lambda x: ' '.join(x))
df['text'] = df['text'].apply(lambda x: ' '.join(x))
```

## Encoding of subject column

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
subject_encoded = ordinal_encoder.fit_transform(df[['subject']])

# Convert the NumPy array to a DataFrame
df_subject_encoded = pd.DataFrame(subject_encoded, columns=['subject_encoded'])

# Concatenate the ordinal encoded 'subject' with the original DataFrame
df = pd.concat([df, df_subject_encoded], axis=1)

# Drop the original 'subject' column as it's no longer needed
df = df.drop('subject', axis=1)
```

## Pretrained model Bert

```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-cased')
```

## Convert data into numerical form

```python
seq_length=300
tokens=tokenizer(df['text'].tolist(),
                max_length=seq_length,
                truncation=True,
                padding='max_length',
                add_special_tokens=True,
                return_tensors='np')
```

## Labels for bert model

```python
num_samples=len(df)
import numpy as np
arr=df['label'].values
labels = np.zeros((num_samples, arr.max()+1))
labels[np.arange(num_samples), arr] = 1
```

```python
labels
```

```python
df_ids = tokens['input_ids']
df_mask = tokens['attention_mask']
```

## Final dataset for Bert model

```python
import tensorflow as tf
dataset=tf.data.Dataset.from_tensor_slices((df_ids,df_mask,labels))
dataset.take(1)
```

```python
def map_func(input_ids, masks, labels):
    return {'input_ids': input_ids,
           'attention_mask': masks}, labels
```

```python
dataset = dataset.map(map_func)
dataset.take(1)
```

```python
batch_size=8
dataset=dataset.shuffle(10000).batch(batch_size,drop_remainder=True)
```

```python
dataset.take(1)
```

## Train_validation Split

```python
def get_dataset_partitions_tf(ds, train_split=0.8, val_split=0.2, shuffle=True, shuffle_size=10000):
    assert (train_split + val_split) == 1
    
    ds_size = len(ds)
    
    if shuffle:
        ds = ds.shuffle(shuffle_size, seed=12)
    
    train_size = int(train_split * ds_size)
    val_size = int(val_split * ds_size)
    
    train_ds = ds.take(train_size)    
    val_ds = ds.skip(train_size).take(val_size)
    
    return train_ds, val_ds,
```

```python
train_ds, val_ds = get_dataset_partitions_tf(dataset)
```

```python
print(len(train_ds))
print(len(val_ds))
```

# Model

```python
from transformers import TFAutoModel
```

## Fine tune the TFAutoModel

```python
input_ids = tf.keras.layers.Input(shape=(300,), name='input_ids', dtype='int32')
mask = tf.keras.layers.Input(shape=(300,), name='attention_mask', dtype='int32')

bert = TFAutoModel.from_pretrained('bert-base-cased')
# transformer
embeddings = bert.bert(input_ids, attention_mask=mask)[1]

# classifier head
x = tf.keras.layers.Dense(1024, activation='relu')(embeddings)
y = tf.keras.layers.Dense(2, activation='softmax', name='outputs')(x)

model = tf.keras.Model(inputs=[input_ids, mask], outputs=y)

optimizer = tf.keras.optimizers.Adam(learning_rate=1e-5)
loss = tf.keras.losses.CategoricalCrossentropy()
acc = tf.keras.metrics.CategoricalAccuracy('accuracy')

model.compile(optimizer=optimizer, loss=loss, metrics=[acc])

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=1
)
```

#### The Fine tuned model is showing impressive accuracy.
#### It took around 40 minutes for training because the number of points are too high to train.
